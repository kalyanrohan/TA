```
'''
You are excited in starting up a toy store business and want to use the excitement and euphoria of the release of
"Episode X: The Force Sleeps" movie as momentum in launching your store. You have decided in building an
application using the Python programming language.
1. (5 points) Design the class structure to store records of the toys. See the exhibit for sample toys that
you are selling.
2. (10 points) Create a function that will save and append a new toy record to the "toyinventory.txt" file
3. (5 points) When your application runs (is executed), it loads the toy inventory records. Create a function
to do this (load toy records into a list).
4. (5 points) Create a function to search for a particular toy record based on 'Stock Code' in the list. This
function should return the record number (index in the list).
5. (5 points) Create a function to update data (record fields) on a record. Hint: you can use/call the function
created in question 4.
6. (10 points) Create a function "Purchase" that: accepts a Stock Code and quantity as arguments, reduce
the quantity, and return the Name and Price of the item so that in can be printed on the invoice.
7. (5 points) Create a function Display that will display all the items in the toy inventory.
8. (5 points) Create a function Exit that will be executed when the program ends. This function will save
all records to the inventory file and closes any open files.
'''
```

```
records=[]
#1 toy class
class Toys:
    def __init__(self,name,stock_code,manufacturer,price,stock,quantity):
        self.name=name
        self.stock_code=stock_code
        self.manufacturer=manufacturer
        self.price=price
        self.stock=stock
        #for number 6
        self.quantity=quantity

    #2 record data
    def append(self):
        with open('toyinventory.txt','a') as f:
            f.write(f'name={self.name},stock_code={self.stock_code},manufacturer={self.manufacturer},price={self.price},stock={self.stock},quantity={self.quantity}'+'\n')
    

#3 get data   
def get_records(file_name='toyinventory.txt'):
    with open(f'{file_name}','r') as f:
        global records
        records=f.read().split('\n')


toy1=Toys(input("name: "),input("stock_code: "),input("manufacturer: "),input("price: "),input("stock: "),input("quantity: "))
toy2=Toys(input("name: "),input("stock_code: "),input("manufacturer: "),input("price: "),input("stock: "),input("quantity: "))
toy3=Toys(input("name: "),input("stock_code: "),input("manufacturer: "),input("price: "),input("stock: "),input("quantity: "))
toy_list = [toy1,toy2,toy3]

#4 search function
def search(stock_code):
    for toy in toy_list:
        if toy.stock_code == stock_code:
            return toy_list.index(toy)


#5 creating update function
def update(stock_code):
    for toy in toy_list:
        if stock_code==toy.stock_code:
            toy.name=input("name: ")
            toy.stock_code=input("stock_code: ")
            toy.manufacturer=input("manufacturer: ")
            toy.price=input("price: ")
            toy.stock=input("stock: ")
            toy.quantity=input("quantity: ")

#6 update function
def purchase(stock_code,quantity):
    for toy in toy_list:
        if stock_code==toy.stock_code and quantity==toy.quantity:
            toy.quantity-=quantity
            print(f'Name= {toy.name},Price= {toy.price}')
        else:
            print('item not found!')
#7 display function
def display():
    for toy in records:
        print(toy)

#8 exit function
def exit(file_name='toyinventory.txt'):
    with open(f'{file_name}', "w") as f:
        for toys in records:
            f.write(toys+"\n")

#test the functions here
if __name__ == "__main__":
    toy1.append()
    toy2.append()
    toy3.append()
    get_records()
    search(SW0001)
    update(SW0002)
    purchase(SW0003,1)
    display()
    exit()
```
